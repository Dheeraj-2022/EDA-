# Exploratory Data Analysis(EDA)

The dataset was released by Aspiring Minds from the Aspiring Mind Employment Outcome 2015 (AMEO). The study is primarily limited  only to students with engineering disciplines. The dataset contains the employment outcomes of engineering graduates as dependent variables (Salary, Job Titles, and Job Locations) along with the standardized scores from three different areas – cognitive skills, technical skills and personality skills. The dataset also contains demographic features. The dataset  contains  around  40 independent variables and 4000 data points. The independent variables are both continuous and categorical in nature. The dataset contains a unique identifier for each candidate.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

# Univariate Analysis

```python
num_df=df.select_dtypes(include=['int64','float'])
num_df.columns
```

```python
num_df.describe()
```

```python
num_df.plot(kind='hist',subplots=True, figsize=(19, 19),layout=(6,5))
```

```python
num_df.plot(kind='box',subplots=True, figsize=(19, 19),layout=(6,5))
```

```python
num_df.plot(kind='kde',subplots=True, figsize=(19, 19),layout=(6,5))
```

```python
percentile25 = num_df.quantile(0.25)
percentile75 = num_df.quantile(0.75)
    
iqr=percentile75-percentile25
    
    
upper_limit = percentile75 + 1.5 * iqr
lower_limit = percentile25 - 1.5 * iqr
    
print(num_df[num_df > upper_limit])
print(num_df[num_df < lower_limit])
    
    
```

```python
new_df = num_df[num_df < upper_limit]
new_df.shape
```

```python
new_df.plot(kind='box',subplots=True, figsize=(19, 19),layout=(6,5))
```

If we remove the outliers for every numerical data by using IQR method we are losing the maximum data for the                    Columns=[ComputerScience,MechanicalEngg,ElectricalEngg,TelecomEngg,CivilEngg ]
